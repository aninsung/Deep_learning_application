## 1. 라이브러리 임포트 및 데이터 생성

```python
# 랜덤 데이터 생성 
n_samples = 100  # 샘플 수를 100으로 설정
n_timesteps = 10  # 시퀀스 길이
n_features = 5    # 특성 수
x_train = np.random.rand(n_samples, n_timesteps, n_features).astype(np.float32)
y_train = np.random.randint(0, 2, size=(n_samples, 1)).astype(np.float32)  # 이진 레이블

n_test_samples = 20  # 테스트 샘플 수를 20으로 설정
x_test = np.random.rand(n_test_samples, n_timesteps, n_features).astype(np.float32)
y_test = np.random.randint(0, 2, size=(n_test_samples, 1)).astype(np.float32)

```

##  2. Transformer Encoder 함수 정의

```python
# Transformer 엔코더 함수 정의
def transformer_encoder(inputs, head_size, num_heads, ff_dim, dropout=0):
    # Attention과 정규화
    x = layers.MultiHeadAttention(key_dim=head_size, num_heads=num_heads, dropout=dropout)(inputs, inputs)
    x = layers.Dropout(dropout)(x)
    x = layers.LayerNormalization(epsilon=1e-6)(x + inputs)  # residual connection
    
    # Feed Forward Network
    x_ff = layers.Dense(ff_dim, activation="relu")(x)
    x_ff = layers.Dropout(dropout)(x_ff)
    x = layers.Dense(inputs.shape[-1])(x_ff)  # 변환 후 입력 차원으로 되돌림
    x = layers.LayerNormalization(epsilon=1e-6)(x + inputs)  # residual connection
    return x
```

## 3. 모델 구축 함수 정의

```python
# 모델 구축 함수
def build_model(input_shape, head_size, num_heads, ff_dim, num_transformer_blocks, mlp_units, dropout=0, mlp_dropout=0):
    inputs = keras.Input(shape=input_shape)
    x = inputs
    for _ in range(num_transformer_blocks):
        x = transformer_encoder(x, head_size, num_heads, ff_dim, dropout)

    x = layers.GlobalAveragePooling1D(data_format="channels_last")(x)
    for dim in mlp_units:
        x = layers.Dense(dim, activation="relu")(x)
        x = layers.Dropout(mlp_dropout)(x)
    
    outputs = layers.Dense(1, activation="sigmoid")(x)  # 이진 분류를 위한 수정
    return keras.Model(inputs, outputs)
```

## 4. 모델 컴파일 및 훈련

```python
# 모델 구축
input_shape = x_train.shape[1:]

model = build_model(
    input_shape,
    head_size=256,
    num_heads=4,
    ff_dim=4,
    num_transformer_blocks=4,
    mlp_units=[128],
    mlp_dropout=0.4,
    dropout=0.25,
)

# 모델 컴파일
model.compile(
    loss="binary_crossentropy",  # 이중 분류에 맞는 손실 함수
    optimizer=keras.optimizers.Adam(learning_rate=1e-4),
    metrics=["accuracy"],
)

model.summary()

# 모델 훈련
callbacks = [keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)]
model.fit(
    x_train,
    y_train,
    validation_split=0.2,
    epochs=20,  # 에포크 수를 줄였습니다.
    batch_size=32,  # 배치 크기를 32로 설정했습니다.
    callbacks=callbacks,
)

# 모델 평가
model.evaluate(x_test, y_test, verbose=1)
```
